# Missing Values

```python
missing = df.isna().sum()
missing_pct = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    "Missing Count": missing,
    "Missing %": missing_pct
})

display(missing_df[missing_df["Missing Count"] > 0])
```

# Fill Missing Values

```python
fill_map = {
    "Value_Deal": "None",
    "Multiple_Lines" : "No",
    "Internet_Type" : "No",
    "Online_Security" : "No",
    "Online_Backup" : "No",
    "Device_Protection_Plan": "No",
    "Premium_Support": "No",
     "Streaming_TV": "No",
     "Streaming_Movies": "No",
     "Streaming_Music": "No",
      "Unlimited_Data": "No",
      "Churn_Category": "other",
     "Churn_Reason": "other"
}

df.fillna(value=fill_map, inplace=True)

df.isna().sum()  # verify result
```
